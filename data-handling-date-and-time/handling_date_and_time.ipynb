### As we can see the data type of date columns in both datasets is object, need to convert it to date data type

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])

date.info()
```

####1. Extract Year from Date

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

####2. Extract Month Number from Date

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

### 3. Extract Month Name from Date

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

### 4. Extract Day from Date

```python
date['date_day'] = date['date'].dt.day

date.head()
```

### 5. Extract Day of the Week (starts from Monday)

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

### 6. Extract Day Name 

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.head()
```

### 7. Check if the day is weekend

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.head()
```

### 8. Extract Week Number of the Calender Year

```python
date['date_week'] = date['date'].dt.week

date.head()
```

### 9. Extract Quarter Number of the Calender Year



```python
date['quarter'] = date['date'].dt.quarter

date.head()
```

### 10. Extract Semester Number of the Calender Year

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.head()
```

### Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
# Extract Only Days and not time
(today - date['date']).dt.days
```

### Extract Months passed by in between dates

```python
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
