### 2. K-1 OneHotEncoding

```python
# Here we delete first encoded value which helps to prevent multicollinearity
# Pandas does not stores the positions and indexes of categories which are encoded and which are dropped
# So, for data analysis we can use pandas OHE and drop first feature but for a machine learning project 
# it is better to use scikit learn's 'OneHotEncoder' class.

pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

### 3. OneHotEncoding using Scikit Learn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
# drop parameter will delete the first category of OHE
# sparse parameter will create a numpy array instead of sparse matrix
# dtype will decide the data type of the encoding if it will be float or integer.
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

### 4. OneHotEncoding with Top Categories

```python
print('Number of Unique Car Brands:', df['brand'].nunique())
df['brand'].value_counts()
```

```python
counts = df['brand'].value_counts()

df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
# This techniques performs OHE for those cars which have value greater than 100 or more, all the other brands are labelled as 'Others' 
pd.get_dummies(df['brand'].replace(repl, 'others')).sample(5)
```
